```python
clustering = AgglomerativeClustering(distance_threshold=None, n_clusters=3)#AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')

predicted_clusters = clustering.fit_predict(data[:1000])
predicted_clusters
```

```python
def plot_clustering(predicted_clusters, data):
    print(data.shape)
    print(predicted_clusters.shape)
    plt.scatter(data[predicted_clusters ==0,0], data[predicted_clusters == 0,1], s=100, c='red')
    plt.scatter(data[predicted_clusters==1,0], data[predicted_clusters == 1,1], s=100, c='black')
    plt.scatter(data[predicted_clusters ==2,0], data[predicted_clusters == 2,1], s=100, c='blue')
    plt.scatter(data[predicted_clusters ==3,0],data[predicted_clusters == 3,1], s=100, c='cyan')
    plt.show()
```

```python
plot_clustering(predicted_clusters, data[:1000].to_numpy())
```
